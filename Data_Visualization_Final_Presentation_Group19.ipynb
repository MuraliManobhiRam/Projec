```python
# @title Plotting The Histogram For the Rice Area,Production And Wheat Area,Production
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Here we are plotting rice area distribution utlizing histogram
plt.subplot(2, 2, 1)
sns.histplot(Agriculture_Indian["RICE AREA (1000 ha)"], bins=30, kde=True, color='blue')
plt.title("Distribution Of Area For Rice")

# Here we are plotting rice production distribution utlizing histogram
plt.subplot(2, 2, 2)
sns.histplot(Agriculture_Indian["RICE PRODUCTION (1000 tons)"], bins=30, kde=True, color='salmon')
plt.title("Distribution of Production for Rice")

# Here we are plotting wheat area distribution utlizing histogram
plt.subplot(2, 2, 3)
sns.histplot(Agriculture_Indian["WHEAT AREA (1000 ha)"], bins=30, kde=True, color='yellow')
plt.title("Distribution Of Area For Wheat")

# Here we are plotting wheat production distribution utlizing histogram
plt.subplot(2, 2, 4)
sns.histplot(Agriculture_Indian["WHEAT PRODUCTION (1000 tons)"], bins=30, kde=True, color='pink')
plt.title("Distribution of Production for Wheat")

plt.tight_layout()
plt.show()
```

```python
# @title Plotting The BoxPlot For the Rice,Sugarcane,Wheat Area across different states
# Here we are setting the colors
colors = ["#FF9247", "#1E40FF", "#12DD42"]
# Here we are setting the seaborn as style
sns.set_style("whitegrid")


# Plot the box plots with for the rice area and different states
# Here we are setting the figure size
plt.figure(figsize=(12, 8))
sns.boxplot(x='State Name', y='RICE AREA (1000 ha)', data=Agriculture_Indian, color=colors[0])
plt.title('Distribution of Rice Area across Different States')
plt.xticks(rotation=90)
plt.ylabel('Rice Area (1000 ha)')
plt.xlabel('State Name')
plt.show()
# Plot the box plots with for the rice area and different states
# Here we are setting the figure size
plt.figure(figsize=(12, 8))
sns.boxplot(x='State Name', y='WHEAT AREA (1000 ha)', data=Agriculture_Indian, color=colors[1])
plt.title('Distribution of Wheat Area across Different States')
plt.xticks(rotation=90)
plt.ylabel('Wheat Area (1000 ha)')
plt.xlabel('State Name')
plt.show()
# Plot the box plots with for the sugarcane area and different states
# Here we are setting the figure size
plt.figure(figsize=(12, 8))
sns.boxplot(x='State Name', y='SUGARCANE AREA (1000 ha)', data=Agriculture_Indian, color=colors[2])
plt.title('Distribution of Sugarcane Area across Different States')
plt.xticks(rotation=90)
plt.ylabel('Sugarcane Area (1000 ha)')
plt.xlabel('State Name')
plt.show()
```

```python
# @title Plotting The HistPlot For The RICE,PEARL MILLET Area,Production for The Parameters
# Here We are defining the parameters to plot
Agriculture_Indian_parameters = ['RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)',
                                'PEARL MILLET AREA (1000 ha)', 'PEARL MILLET PRODUCTION (1000 tons)']

# Here we are creating subplots for every parameter
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Here we are flatten the axes array to perform easy iteration
axes = axes.flatten()

# Here we are plotting the histograms for each parameter
for i, Parameters_Agriculture in enumerate(Agriculture_Indian_parameters):
    ax = axes[i]
    ax.hist(Agriculture_Indian[Parameters_Agriculture], bins=40, color='violet', edgecolor='black')
    ax.set_title(Parameters_Agriculture)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# @title Plotting The Scatterplot For The Rice,Wheat,Maize,Sugarcane Area,Production for The Parameters
# Here We are defining the parameters to plot
Agriculture_Indian_parameters = [('RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)'),
                   ('WHEAT AREA (1000 ha)', 'WHEAT PRODUCTION (1000 tons)'),
                   ('MAIZE AREA (1000 ha)', 'MAIZE PRODUCTION (1000 tons)'),
                   ('SUGARCANE AREA (1000 ha)', 'SUGARCANE PRODUCTION (1000 tons)')]

# Here we are creating subplots for every parameter
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Here we are flatten the axes array to perform easy iteration
axes = axes.flatten()

# Here we are plotting the scatterplot for each parameter
for i, pair in enumerate(Agriculture_Indian_parameters):
    ax = axes[i]
    ax.scatter(Agriculture_Indian[pair[0]], Agriculture_Indian[pair[1]], color='green', alpha=0.7)
    ax.set_title(f"{pair[1]} vs {pair[0]}")
    ax.set_xlabel(pair[0])
    ax.set_ylabel(pair[1])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# @title Plotting The Histogram For The Minor Pulses,Wheat,GroundNut,Sugarcane Area,Production for The Parameters
#  Here We are defining the parameters to plot the histogram
Agriculture_Indian_parameters = ['MINOR PULSES AREA (1000 ha)', 'WHEAT AREA (1000 ha)', 'GROUNDNUT AREA (1000 ha)', 'SUGARCANE AREA (1000 ha)']

# Here we are creating subplots for every parameter
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Here we are flatten the axes array to perform easy iteration
axes = axes.flatten()

# Here we are plotting the scatterplot for each parameter
for i, Parameters_Agriculture in enumerate(Agriculture_Indian_parameters):
    ax = axes[i]
    ax.hist(Agriculture_Indian[Parameters_Agriculture], bins=20, color='blue', alpha=0.8)
    ax.set_title(f"Distribution of {Parameters_Agriculture}")
    ax.set_xlabel(Parameters_Agriculture)
    ax.set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# @title Plotting The Trends For The Minor Pulses,Wheat,GroundNut,Sunflower Area,Production for The Parameters
# # Here We are defining the parameters to plot the histogram
#  Here We are defining the parameters to plot the treds for the production
Agriculture_Indian_parameters = ['MINOR PULSES PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)', 'OILSEEDS PRODUCTION (1000 tons)', 'SUNFLOWER PRODUCTION (1000 tons)']

# Group the data by year and calculate the total production for each year
Agriculture_Indian_Production_Over_Time = Agriculture_Indian.groupby('Year')[Agriculture_Indian_parameters].sum()

# Here we are plotting the trends for year
plt.figure(figsize=(12, 8))
for Parameters_Agriculture in Agriculture_Indian_parameters:
    plt.plot(Agriculture_Indian_Production_Over_Time.index, Agriculture_Indian_Production_Over_Time[Parameters_Agriculture], label=Parameters_Agriculture.split(' ')[0])

plt.title("Agricultural Production Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# @title Plotting The HistPlot for The Population, % of Total Population, Literacy Rate (%)
# Visualize the distribution of numerical variables using histograms
Country_Global_Columns = ['Population', '% of Total Population', 'Literacy Rate (%)']
Country_Global[Country_Global_Columns].hist(figsize=(12, 8))
plt.suptitle("Distribution For Poulation , Percentage Of Total Population, Literacy Rate")
plt.show()
```

```python
# @title Plotting The BarChart for Male,Female ratio across different states
# Here We are analyzing the sex ratio across different states
plt.figure(figsize=(12, 8))
# In the column of sex ratio there are commas so i am removing he commas and changing it into integer
Country_Global['Sex Ratio'] = Country_Global['Sex Ratio'].str.replace(',', '').astype(int)

# Utilizing The Bar Chart We are Plotting the sex ratio average across different states
Country_Global.groupby('State')['Sex Ratio'].mean().sort_values().plot(kind='barh', color='green')
plt.title("Average Sex Ratio Across Different States")
plt.xlabel("Sex Ratio")
plt.ylabel("State")
plt.show()
```

```python
# @title Plotting The BarChart for density of population across states which are different
# Explore the population density across different states
plt.figure(figsize=(12, 8))
# In the column of Population Density there are commas so i am removing he commas and changing it into integer
Country_Global['Density (1/km*km)'] = Country_Global['Density (1/km*km)'].str.replace(',', '').astype(int)
# Utilizing The Bar Chart We are Plotting the population density across different states
Country_Global.groupby('State')['Density (1/km*km)'].mean().sort_values().plot(kind='barh', color='orange')
plt.title("Average Population Density Across Different States")
plt.xlabel("Density (1/km*km)")
plt.ylabel("State")
plt.show()
```

```python
# @title Plotting The BarChart for growth decade percentage across states which are different
# Here we are exploring the decadal growth percentage across different states
plt.figure(figsize=(12, 8))
# In the column of Decadal Growth we are removing special characters and cchanging the 'Decadal Growth (%)' column to float
Country_Global['Decadal Growth (%)'] = Country_Global['Decadal Growth (%)'].str.replace('−', '-').str.rstrip('%').astype(float)

# Utilizing The Bar Chart We are Plotting the average decadal growth across different states
Country_Global.groupby('State')['Decadal Growth (%)'].mean().sort_values().plot(kind='barh', color='pink')
plt.title("Average Decadal Growth Percentage Across Different States")
plt.xlabel("Decadal Growth Percentage")
plt.ylabel("State")
plt.show()
```

```python
# @title Plotting The Corelaion Matrix for the data which is global country population
# In this case we are exclude non-numeric columns and calculating the correlation matrix
Country_Global_Numeric_Columns = Country_Global.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = Country_Global[Country_Global_Numeric_Columns].corr()

# Utilizing the Heat Map we are plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix For the Global Country Data")
plt.show()
```

```python
# @title Merging The Both Agriculuture and Global Country Data
# Here performing inner join and merging the both the datasets on the column which is "state"
Agriculture_Global_Merged = pd.merge(Agriculture_Indian, Country_Global, left_on='State Name', right_on='State', how='inner')

# Here We are Displaying the first rows of the dataset which is merged
print(Agriculture_Global_Merged.head())
```

```python
# @title Printing the Columns For The Merged Data
# Printing the columns For both Agriculture and Global Country
Agriculture_Global_Merged.columns
```

```python
# @title Checking The DataTypes
# Using the dtypes command we are checking the types of merged data.
Agriculture_Global_Merged.dtypes
```

```python
# @title Plotting The BarChart for disturbution of population across states which are different
# Here We are Converting the  population related columns to types which are numeric
Agriculture_Global_Merged['Population'] = Agriculture_Global_Merged['Population'].str.replace(',', '').astype(int)
Agriculture_Global_Merged['Males'] = Agriculture_Global_Merged['Males'].str.replace(',', '').astype(int)
Agriculture_Global_Merged['Females'] = Agriculture_Global_Merged['Females'].str.replace(',', '').astype(int)
Agriculture_Global_Merged['Rural Population'] = Agriculture_Global_Merged['Rural Population'].str.replace(',', '').astype(int)
Agriculture_Global_Merged['Urban Population'] = Agriculture_Global_Merged['Urban Population'].str.replace(',', '').astype(int)
# Utilizing the Barchart we are plotting the across state population disturbution
plt.figure(figsize=(12, 8))
Agriculture_Global_Merged.groupby('State Name')['Population'].sum().sort_values().plot(kind='barh', color='red')
plt.title("Population Distribution Across States")
plt.xlabel("Population")
plt.ylabel("State")
plt.show()
```

```python
# @title Plotting The ScatterPlot for disturbution of population across states and agricultural production
# Here we are calculating total agricultural production
Agriculture_Global_Merged['Total Crop Area (1000 ha)'] = Agriculture_Global_Merged.filter(like='AREA').mean(axis=1)
Agriculture_Global_Merged['Total Crop Production (1000 tons)'] = Agriculture_Global_Merged.filter(like='PRODUCTION').sum(axis=1)
Agriculture_Global_Merged['Average Crop Yield (Kg per ha)'] = Agriculture_Global_Merged.filter(like='YIELD').mean(axis=1)
# Utilizing the scatterplot we are plotting the across state population disturbution and agriculture production
plt.figure(figsize=(12, 8))
plt.scatter(Agriculture_Global_Merged['Population'], Agriculture_Global_Merged['Total Crop Area (1000 ha)'], color='violet', label='Crop Area')
plt.scatter(Agriculture_Global_Merged['Population'], Agriculture_Global_Merged['Total Crop Production (1000 tons)'], color='yellow', label='Crop Production')
plt.scatter(Agriculture_Global_Merged['Population'], Agriculture_Global_Merged['Average Crop Yield (Kg per ha)'], color='orange', label='Crop Yield')
plt.xlabel('Population')
plt.ylabel('Agricultural Production')
plt.title('Population Distribution Across States vs. Agricultural Production')
plt.legend()
plt.show()
```

```python
# @title Plotting The ScatterPlot for land usage for agriculture and trends for urbanization
# Here we are setting the Figure Size
plt.figure(figsize=(12, 8))
plt.scatter(Agriculture_Global_Merged['Urban Population'], Agriculture_Global_Merged['Total Crop Area (1000 ha)'], color='blue', label='Crop Area')
plt.scatter(Agriculture_Global_Merged['Urban Population'], Agriculture_Global_Merged['Total Crop Production (1000 tons)'], color='grey', label='Crop Production')
plt.scatter(Agriculture_Global_Merged['Urban Population'], Agriculture_Global_Merged['Average Crop Yield (Kg per ha)'], color='pink', label='Crop Yield')
# Here we are setting xlabel
plt.xlabel('Urban Population')
# Here we are setting ylabel
plt.ylabel('Agricultural Land Usage')
#Here we are setting the title utilizing the title command
plt.title('Urbanization Trends vs. Agricultural Land Usage')
#Here we are setting the legend
plt.legend()
#Utilizing the show command we are displaying the plot
plt.show()
```

```python
# @title Plotting The ScatterPlot for productivity of agriculture and density of population
# Here we are setting the Figure Size
plt.figure(figsize=(12, 8))
plt.scatter(Agriculture_Global_Merged['Density (1/km*km)'], Agriculture_Global_Merged['Total Crop Production (1000 tons)'], color='purple', label='Crop Production')
plt.scatter(Agriculture_Global_Merged['Density (1/km*km)'], Agriculture_Global_Merged['Average Crop Yield (Kg per ha)'], color='lightgreen', label='Crop Yield')
plt.xlabel('Population Density (1/km*km)')
plt.ylabel('Agricultural Productivity Metrics')
plt.title('Population Density vs. Agricultural Productivity')
plt.legend()
plt.show()
```
